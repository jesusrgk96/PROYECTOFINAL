```python
import pandas as pd

df = pd.read_csv('/content/country_vaccinations.csv')
display(df.head())
```

## Preparar y Visualizar Estructura del DataFrame

### Subtask:
Mostrar la estructura y los tipos de datos de cada columna del DataFrame usando `df.info()`. Luego, convertir la columna de fechas a un tipo de datos `datetime64` para permitir operaciones de tiempo adecuadas.


```python
df.info()
df['date'] = pd.to_datetime(df['date'])
print("Date column converted to datetime type.")
```

## Vacunas Aplicadas por Compañía

### Subtask:
Calcular la cantidad total de vacunas aplicadas por cada compañía. Esto se logrará agrupando el DataFrame por la columna `vaccines` y sumando las `total_vaccinations`.


```python
total_vaccines_by_company = df.groupby('vaccines')['total_vaccinations'].sum().sort_values(ascending=False)
print("Total vaccines applied by company:")
display(total_vaccines_by_company)
```

## Total de Vacunas Aplicadas Mundialmente

### Subtask:
Determinar la cantidad total de vacunas aplicadas en todo el mundo sumando la columna `total_vaccinations`.


```python
total_worldwide_vaccinations = df['total_vaccinations'].sum()
print(f"Total worldwide vaccinations: {total_worldwide_vaccinations:.2f}")
```

## Promedio de Vacunas por País

### Subtask:
Calcular el promedio de vacunas aplicadas por cada país, agrupando por `country` y obteniendo el promedio de `total_vaccinations`.


```python
avg_vaccines_per_country = df.groupby('country')['total_vaccinations'].mean().sort_values(ascending=False)
print("Average total vaccinations per country:")
display(avg_vaccines_per_country)
```

## Vacunas Aplicadas en una Fecha Específica

### Subtask:
Obtener la cantidad total de vacunas aplicadas en todo el mundo el día 29 de enero de 2021, filtrando el DataFrame por la fecha y sumando las `daily_vaccinations`.


```python
vaccinations_jan_29 = df[df['date'] == '2021-01-29']['daily_vaccinations'].sum()
print(f"Total worldwide daily vaccinations on 2021-01-29: {vaccinations_jan_29:.2f}")
```

## Crear DataFrame 'conDiferencias'

### Subtask:
Crear un nuevo DataFrame llamado `conDiferencias` que contenga todos los datos originales, más una nueva columna llamada `diferencias`. Esta columna contendrá la diferencia entre las columnas `daily_vaccinations` y `daily_vaccinations_raw`.


```python
conDiferencias = df.copy()
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
print("DataFrame 'conDiferencias' created with new 'diferencias' column:")
display(conDiferencias.head())
```

## Período de Tiempo de los Registros

### Subtask:
Calcular el período de tiempo (días, meses, años) entre el registro con la fecha más reciente y el registro con la fecha más antigua en el DataFrame.


```python
min_date = df['date'].min()
max_date = df['date'].max()
time_span = max_date - min_date

print(f"Earliest record date: {min_date}")
print(f"Latest record date: {max_date}")
print(f"Time span of records: {time_span}")
```

## Crear DataFrame 'conCantidad'

### Subtask:
Crear un nuevo DataFrame llamado `conCantidad` que incluya todos los datos originales y una nueva columna llamada `canVac`. Esta columna contendrá la cantidad de tipos de vacunas utilizadas en cada registro, obtenida dividiendo la columna `vaccines` por la coma (`,`) y contando los elementos.


```python
conCantidad = df.copy()
conCantidad['canVac'] = conCantidad['vaccines'].apply(lambda x: len(str(x).split(', ')))
print("DataFrame 'conCantidad' created with new 'canVac' column:")
display(conCantidad.head())
```

## Crear DataFrame 'antes20'

### Subtask:
Crear un nuevo DataFrame llamado `antes20` que contenga todos los registros del DataFrame original `df` cuya fecha sea anterior al 20 de diciembre de 2020.



```python
antes20 = df[df['date'] < '2020-12-20']
print("DataFrame 'antes20' created with records before 2020-12-20:")
display(antes20.head())
```

## Crear DataFrame 'pfizer'

### Subtask:
Crear un nuevo DataFrame llamado `pfizer` que contenga todos los registros del DataFrame original `df` donde la columna `vaccines` indique que se utilizó la vacuna 'Pfizer'.



```python
pfizer = df[df['vaccines'].str.contains('Pfizer', na=False)]
print("DataFrame 'pfizer' created with records where 'Pfizer' vaccine was used:")
display(pfizer.head())
```

## Exportar DataFrames a Excel

### Subtask:
Exportar los DataFrames `conDiferencias`, `conCantidad`, `antes20`, y `pfizer` a un archivo Excel llamado `resultadosReto.xlsx`. Cada DataFrame debe estar en una hoja separada dentro del archivo Excel.



```python
output_file = 'resultadosReto.xlsx'
with pd.ExcelWriter(output_file) as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
print(f"DataFrames exported to {output_file}")
```
